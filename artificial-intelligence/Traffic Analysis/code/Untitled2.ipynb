```python
df= pd.read_csv(r"C:\Users\USER\Chameleon\MOP-Code\artificial-intelligence\Traffic Analysis\dataset\cleaned_merged_data.csv")
df
```

```python
# Define the columns that represent different types of vehicles
vehicle_columns = [
    'Light_Vehicles', 'Light_Vehicles_with_Trailer', 'Two_Axle_Trucks_Buses',
    'Three_Axle_Trucks_Buses', 'Four_Axle_Trucks', 'Three_Axle_Articulated_Vehicles',
    'Four_Axle_Articulated_Vehicles', 'Five_Axle_Articulated_Vehicles',
    'Six_Axle_Articulated_Vehicles', 'B_Double_Trucks', 'Double_Road_Trains',
    'Triple_Road_Trains', 'Unknown_Vehicles', 'Motorcycles', 'Bicycles'
]
# Calculate total traffic volume
df['Total_Traffic_Volume'] = df[vehicle_columns].sum(axis=1)
df[['date', 'road_name', 'location', 'suburb', 'time', 'Total_Traffic_Volume']]
```

```python
heavy_vehicle_columns = [
    'Two_Axle_Trucks_Buses', 'Three_Axle_Trucks_Buses', 'Four_Axle_Trucks',
    'Three_Axle_Articulated_Vehicles', 'Four_Axle_Articulated_Vehicles',
    'Five_Axle_Articulated_Vehicles', 'Six_Axle_Articulated_Vehicles',
    'B_Double_Trucks', 'Double_Road_Trains', 'Triple_Road_Trains'
]
# Calculate total heavy vehicle volume
df['Heavy_Vehicle_Volume'] = df[heavy_vehicle_columns].sum(axis=1)
df[['date', 'road_name', 'location', 'suburb', 'time', 'Total_Traffic_Volume', 'Heavy_Vehicle_Volume']]
```

```python

# Calculate heavy vehicle percentagedata
df['Heavy_Vehicle_Percentage'] = (df['Heavy_Vehicle_Volume'] / df['Total_Traffic_Volume']) * 100
# Display the first few rows of the dataset with the new columns
df[['date', 'road_name', 'location', 'suburb', 'time', 'Total_Traffic_Volume', 'Heavy_Vehicle_Volume', 'Heavy_Vehicle_Percentage']]
```

```python
# Explore the speed-related columns in the dataset
speed_columns = ['speed_limit', 'average_speed', '85th_percentile_speed', 'maximum_speed']
speed_data = df[speed_columns].describe()

speed_data
```

```python
def categorize_speeding(row):
    speed_limit = row['speed_limit']
    max_speed = row['maximum_speed']
    
    if max_speed <= speed_limit:
        return 'Normal Speed'
    elif speed_limit < max_speed <= speed_limit * 1.1:
        return 'Low-Level Speeding'
    elif speed_limit * 1.1 < max_speed <= speed_limit * 1.2:
        return 'Moderate Speeding'
    else:
        return 'High-Level Speeding'
```

```python
df['Speeding_Category'] = df.apply(categorize_speeding, axis=1)
df[['speed_limit', 'maximum_speed', 'Speeding_Category']]
```

```python
# Step 4: Track Speeding Incidents
speeding_counts = df['Speeding_Category'].value_counts(normalize=True) * 100
speeding_counts
```

```python
# Step 4: Analyze Speeding Incidents by Suburb
speeding_by_suburb = df.groupby('suburb')['Speeding_Category'].value_counts(normalize=True).unstack().fillna(0)
speeding_by_suburb
```

```python
# Step 5: Analyze Speeding Incidents by Time of Day
speeding_by_time = df.groupby('time')['Speeding_Category'].value_counts(normalize=True).unstack().fillna(0)
speeding_by_time
```
